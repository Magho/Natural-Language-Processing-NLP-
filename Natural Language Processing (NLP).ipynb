```python
# Chunking with NLTK

train_text = state_union.raw('2005-GWBush.txt')
sample_text = state_union.raw('2006-GWBush.txt')

# Now that we have some text, we can train the PunktSentenceTokenizer
custome_sent_tokenizer = PunktSentenceTokenizer(train_text)

# now use this to tokenize the sample text
tokenized = custome_sent_tokenizer.tokenize(sample_text)

# Define a function that will tag each tokenized word with a part of the speech
def process_content():
    try:
        firstTime = False
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            
            # combine the part of speech tag with a regular expression
            chunkGram = r"""Chunk: {<RB.?>*<VB.?>*<NNP>+<NN>?}"""
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            
            # draw the nltk tree
            for subtree in chunked.subtrees(filter=lambda t :t.label() == 'Chunk'):
                print (subtree)
            
            # draw the chunks with nltk
            if firstTime == False:    
                chunked.draw()
                firstTime = True
            
    except Exception as e:
        print(str(e))
        
process_content()
```

```python
import random 
import nltk
from nltk.corpus import movie_reviews
```

##  bulid a model as an appllication 

```python
# build a list of documents
documents = [(list(movie_reviews.words(fileid)), category)
            for category in movie_reviews.categories()
            for fileid in movie_reviews.fileids(category)]

# shuffle the documents 
random.shuffle(documents)

print('Number of Documents:{}'.format(len(documents)))
print ('First Review :{}'.format(documents[0]))

all_words = []
for w in movie_reviews.words():
    all_words.append(w.lower())

all_words = nltk.FreqDist(all_words)

print("Most common words : {}".format(all_words.most_common(15)))
print('The word happy:{}'.format(all_words["happy"]))
```

```python
print(len(all_words))
```

```python
# we 'll use half the words as features
word_features = list(all_words.keys())[:20000]
```

```python
# Build a find_features function that will determine which of the 4000 word features are contained  in a review
def find_features(document):
    words = set(document)
    features = {}
    
    for w in word_features :
        features[w] = (w in words)
        
    return features

# lets use an example of a negative review 
features = find_features(movie_reviews.words('neg/cv000_29416.txt'))
for key, value in features.items():
    if value == True:
        print key
```

```python
print(features)
```

```python
# now lets do it for all the documents 
featuresets = [(find_features(rev), category) for (rev, category) in documents]
```

```python
# we can split the featuresets into training and testing datasets using sklearn
from sklearn import model_selection

# define a seed for reproducability
seed = 1

# split the data into training and testing datasets
training, testing = model_selection.train_test_split(featuresets, test_size = 0.25, random_state = seed)
```

```python
print (len(training))
print (len(testing))
```

```python
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.svm import SVC
```

```python
model = SklearnClassifier(SVC(kernel='linear'))
```

```python
# train the model on the training data
model.train(training)
```

```python
# test on the testing dataset
accuracy = nltk.classify.accuracy(model, testing)
print ('SVC Accuracy : {}'.format(accuracy))
```
